```python
import IPython.display as ipd
import librosa
import librosa.display
import matplotlib.pyplot as plt
import os
import numpy as np
import scipy.io.wavfile as wavfile
# from scipy.io.wavfile import write
```

```python
# Set file path

dataDirectory = './data'
sampleType = 'vocal_audio'
audioName = 'vocals'

filePath = os.path.join(dataDirectory, sampleType, str(audioName) + '.mp3')
```

```python
# Listen audio 
ipd.Audio(filePath)
```

```python
# Retrieve audio data
x, sampleRate = librosa.load(filePath)
```

```python
# Visualize audio data
plt.figure(figsize=(18, 4))
librosa.display.waveplot(x, sr=sampleRate)
```

```python
# Show some specs
data = x
print('Sample bits: ' + str(len(data)) + ' (bits)')
print('Sampling rate: ' + str(sampleRate) + ' (bits/second)')
print('Duration: ' + str(len(data)/sampleRate) + ' (seconds)')
```

```python
def timeToIndex(seconds, sr):
    return np.floor(sr*seconds).astype(int)

def getInterval(s0,s1,Fs):
    i0 = timeToIndex(s0, Fs)
    i1 = timeToIndex(s1, Fs)
    return (i0,i1)
    
def extractInterval(data,s0,s1,Fs):
    interval = getInterval(s0,s1,Fs)
    extractedData = data[interval[0]:interval[1]]
    return extractedData
    
def saveToWav(data,sr,fileName):
    # https://docs.scipy.org/doc/scipy/reference/generated/scipy.io.wavfile.write.html
    audioFilePath = os.path.join(dataDirectory, sampleType, fileName + '.wav')
    wavfile.write(audioFilePath, sr, data)
    return audioFilePath
    

vocalA = extractInterval(x,0,0.9,sampleRate)
vocalE = extractInterval(x,0.9,1.7,sampleRate)
vocalI = extractInterval(x,1.7,2.5,sampleRate)
vocalO = extractInterval(x,2.5,3.3,sampleRate)
vocalU = extractInterval(x,3.3,4.2,sampleRate)

librosa.display.waveplot(vocalA, sr=sampleRate)
librosa.display.waveplot(vocalE, sr=sampleRate)
librosa.display.waveplot(vocalI, sr=sampleRate)
librosa.display.waveplot(vocalO, sr=sampleRate)
librosa.display.waveplot(vocalU, sr=sampleRate)
    
```

```python
vocalAPath = saveToWav(vocalA,sampleRate,'vocalA')
vocalEPath = saveToWav(vocalE,sampleRate,'vocalE')
vocalIPath = saveToWav(vocalI,sampleRate,'vocalI')
vocalOPath = saveToWav(vocalO,sampleRate,'vocalO')
vocalUPath = saveToWav(vocalU,sampleRate,'vocalU')
```

```python
ipd.Audio(vocalAPath)
```
